```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_excel('supermarket_sales.xlsx')
dataset.head()
```

```python
dataset.describe()
```

```python
plt.hist(dataset['Total'], bins=10)  # Adjust 'bins' as needed
plt.xlabel('Total')
plt.ylabel('Product line')
plt.title('Histogram of Total Sales')
plt.show()
```

```python
# 1. Histogram of Total Sales
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.hist(dataset['Total'], bins=10)
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Total Sales')
plt.show()
```

```python
# 2. Bar chart of Product Line counts
plt.figure(figsize=(10, 6))
product_line_counts = dataset['Product line'].value_counts()
plt.bar(product_line_counts.index, product_line_counts.values)
plt.xlabel('Product Line')
plt.ylabel('Number of Sales')
plt.title('Number of Sales per Product Line')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# 3. Scatter plot of Quantity vs. Total
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Quantity'], dataset['Total'])
plt.xlabel('Quantity')
plt.ylabel('Total Sales')
plt.title('Quantity vs. Total Sales')
plt.show()
```

```python
# 4. Box plot of Total Sales by Gender
plt.figure(figsize=(8, 6))
dataset.boxplot(column='Total', by='Gender')
plt.ylabel('Total Sales')
plt.title('Total Sales by Gender')
plt.suptitle('') # Remove the default boxplot title
plt.show()
```

```python

# 5. Pie chart of Payment Method distribution
plt.figure(figsize=(8, 8))
payment_counts = dataset['Payment'].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Payment Methods')
plt.show()
```

```python
# 6. Line plot of Total Sales over time (assuming you have a date column)
# Replace 'Date' with the actual name of your date column
if 'Date' in dataset.columns:
    plt.figure(figsize=(12, 6))
    dataset.groupby('Date')['Total'].sum().plot()  # Sum of total sales per date
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.title('Total Sales over Time')
    plt.show()
else:
    print("No 'Date' column found in the dataset. Skipping the time series plot.")
```
