```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/Admission_Predict.csv')

df.head()

```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['GRE Score'], df['CGPA'])
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('GRE Score vs. CGPA')
plt.show()
```

```python
# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Chance of Admit '])
plt.xlabel('Applicant Index')
plt.ylabel('Chance of Admit')
plt.title('Chance of Admit for Applicants')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(8, 6))
df['Research'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Research Experience Distribution')
plt.ylabel('')  # Hide y-label for pie chart
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['Research'] == 0]['Chance of Admit '],
             df[df['Research'] == 1]['Chance of Admit ']],
            labels=['No Research', 'Research'])
plt.ylabel('Chance of Admit')
plt.title('Chance of Admit based on Research Experience')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
df['University Rating'].value_counts().plot(kind='bar')
plt.xlabel('University Rating')
plt.ylabel('Number of Applicants')
plt.title('Distribution of University Ratings')
plt.show()
```
